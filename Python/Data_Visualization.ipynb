```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
# load & read the dataset.
df = pd.read_excel('heatmap_data.xlsx')
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
lb = LabelEncoder()
```

```python
df['gender'] = lb.fit_transform(df['gender'])
```

```python
df['department'] = lb.fit_transform(df['department'])
```

```python
df['Accuracy of Work'] = lb.fit_transform(df['Accuracy of Work'])
```

```python
df['Meeting Project Deadlines'] = lb.fit_transform(df['Meeting Project Deadlines'])
```

```python
df['churned'] = lb.fit_transform(df['churned'])
```

```python
df.dtypes
```

```python
df.corr()
```

```python
plt.figure(figsize = (10,12))
sns.heatmap(df.corr(), annot = True)
plt.show()
```

```python
cor = df.loc[:,['Task Completion Rates','Decision-Making Skills Rating','Sales Revenue Generated']]
```

```python
cor.corr()
```
