```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('courses_data.csv')
df
```

```python
# it shows the datatypes of all columns
df.dtypes
```

```python
# it shows the stastical summary of data
df.describe()
```

```python
freq_churned = df['Credits'].value_counts()

plt.figure(figsize = (4,6))
freq_churned.plot(kind = 'bar')
plt.title('Bar Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
cross_tab = pd.crosstab(df['Course_Name'], df['Credits'])

plt.figure(figsize=(4, 6)) 
cross_tab.plot(kind='bar', stacked=False)
plt.title('Course Name by Credits')
plt.xlabel('Course Name')
plt.ylabel('Count')
plt.show()
```

```python
percentage = df['Credits'].value_counts()

plt.figure(figsize = (4,6))
percentage.plot(kind = 'pie', autopct = '%1.2f%%')
plt.title('Pie Chart')
plt.ylabel('')
plt.show()
```

```python
credit_counts = df['Credits'].value_counts()

plt.figure(figsize=(4, 6))
credit_counts.plot(kind='pie', autopct='%1.2f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Pie Charts')
plt.ylabel('')  # Hide the y-axis label
plt.tight_layout()
plt.show()
```

```python
# Line Plot
plt.figure(figsize = (4,6))
cross_tab.plot( kind = 'line', marker = 'o')
plt.title('Line plot')
plt.xlabel('Course Name')
plt.ylabel('Credits')
plt.show()
```

```python
# Histogram Plot
plt.figure(figsize = (4,6))
df['Credits'].plot(kind = 'hist', bins = 5)
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

sns.histplot(df['Credits'], bins = 5)
```
