```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading CSV File. 
df = pd.read_csv('courses_data.csv')
df
```

```python
# Reading CSV Second File.
data = pd.read_csv('courses_dataa.csv')
data
```

```python
# Joining the 2 file
joined_data = pd.merge(df,data, on = 'Course_ID', how = 'inner')
joined_data.head()
```

```python
# Conate the 2 File.
concatenated_data = pd.concat([df,data])
concatenated_data.head()
```

```python
# Cross_tab:- summarize the relationship between two or more categorical variables.
cross_tab = pd.crosstab(joined_data['Course_Name_x'], joined_data['Course_Name_y'])
cross_tab
```

```python
# Method to calculate frequencies for each category.
df['Course_Name'].value_counts()
```

```python
# Method to group data by a nominal variable on a numeric data.
grouped_var = df.groupby('Course_Name')['Credits'].mean()
grouped_var
```

```python
# It shows the statistical Summary of data.
df.describe()
```

```python
pivot_table = df.pivot_table(index = 'Course_Name',
                             values = ['Course_ID','Credits'],
                            aggfunc = {'Course_Name':'count','Course_ID':'mean','Credits':'mean'})
pivot_table
```

```python
pivot_table = pivot_table.rename(columns = {'Course_Name':'New_Course',
                                            'Course_ID':'Book_id',
                                            'Credits':'Points'})
pivot_table
```
